```python
import pandas
from tqdm.auto import tqdm
import requests
```

```python
data = []
```

```python
len(extract[0])
```

```python
url = "https://www.edinburghmarathon.com/results?event={event}&gender={gender}&page={page}"
event = 1007
for gender in {"M", "F", "N"}:
    for page in tqdm(range(1, 700)):
        data_url = url.format(page=page, gender=gender, event=event)
        #print(data_url)
        extract = pandas.read_html(requests.get(data_url,
                                   headers={'User-agent': 'Mozilla/5.0'}).text)
        if len(extract[0]) == 0:
            break
        data.append(extract[0])
```

```python
frames = pandas.concat(data)
frames = frames[::2]
frames.index = frames['Position (Overall)']
```

```python
frames['Category'] = frames['Position (Category)'].map(lambda x: x.split()[1])
```

```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.mlab import GaussianKDE
import matplotlib.patheffects as pe
import numpy as np
```

```python
def str_to_seconds(time):
    com = time.split(":")
    return int(com[2])+60*int(com[1])+3600*int(com[0])
```

```python
frames['Chip Time (seconds)'] = frames['Chip Time'].map(str_to_seconds)
```

```python
frames.to_csv("edinburgh-marathon-2024.csv")
```

```python
quantiles = pandas.DataFrame()

categories = ("U35", "35", "40", "45", "50", "55", "60", "65", "70")
for category, ax in zip(categories, axes.flatten()):
    men = frames[frames['Category']==f"{category}M"]
    women = frames[frames['Category']==f"{category}F"]
    
    quantiles[f'men club {category}'] = (men[~men['Club'].isna()]['Chip Time (seconds)']/60).quantile(q=np.linspace(0.05,1,20))
    quantiles[f'men non club {category}'] = (men[men['Club'].isna()]['Chip Time (seconds)']/60).quantile(q=np.linspace(0.05,1,20))
    quantiles[f'men overall {category}'] = (men['Chip Time (seconds)']/60).quantile(q=np.linspace(0.05,1,20))

    quantiles[f'women club {category}'] = (women[~women['Club'].isna()]['Chip Time (seconds)']/60).quantile(q=np.linspace(0.05,1,20))
    quantiles[f'women non club {category}'] = (women[women['Club'].isna()]['Chip Time (seconds)']/60).quantile(q=np.linspace(0.05,1,20))
    quantiles[f'women overall {category}'] = (women['Chip Time (seconds)']/60).quantile(q=np.linspace(0.05,1,20))
```
